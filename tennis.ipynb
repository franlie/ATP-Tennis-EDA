```python
#who won the most matches in one season and which year was it
```

```python
%%sql

SELECT YEAR, winner, max(COUNT) as max_count from (
    SELECT Winner, strftime('%Y', Date) as YEAR, COUNT(Winner) as COUNT
    FROM tennis
    GROUP BY YEAR, Winner
    ORDER BY COUNT DESC
)

```

```python
#top ten tournaments won on Grass
```

```python
%%sql

SELECT Winner, Surface, COUNT(*) as COUNT
FROM tennis
WHERE Round = "The Final" and Surface = "Grass"
GROUP BY Winner
ORDER BY COUNT DESC
LIMIT 10
```

```python
#top ten tournaments won this year (2023)
```

```python
%%sql

SELECT Winner, COUNT(*) as COUNT
FROM tennis
WHERE Round = "The Final" and strftime('%Y', Date) = "2023"
GROUP BY Winner
ORDER BY COUNT DESC
LIMIT 10
```

```python
#how many duels between Rafael Nadal and Roger Federer and who won?
```

```python
%%sql

SELECT Date, Player_1, Player_2, Winner
FROM tennis
WHERE (Player_1 LIKE "Nadal R." OR Player_1 LIKE "Federer R.")
AND (Player_2 LIKE"Federer R." OR Player_2 LIKE "Nadal R.")
ORDER BY Date ASC
```

```python
#how many matches between Nadal and Federer and what's the head vs head stats
```

```python
%%sql

select COUNT(*) from (
    SELECT Date, Player_1, Player_2, Winner
    FROM tennis
    WHERE (Player_1 LIKE "Nadal R." OR Player_1 LIKE "Federer R.")
    AND (Player_2 LIKE"Federer R." OR Player_2 LIKE "Nadal R.")
    ORDER BY Date ASC

)
```

```python
%%sql

select COUNT(*) from (
    SELECT Date, Player_1, Player_2, Winner
    FROM tennis
    WHERE (Player_1 LIKE "Nadal R." OR Player_1 LIKE "Federer R.")
    AND (Player_2 LIKE"Federer R." OR Player_2 LIKE "Nadal R.")
    ORDER BY Date ASC

)
WHERE Winner ="Nadal R."
```

```python
%%sql

select COUNT(*) from (
    SELECT Date, Player_1, Player_2, Winner
    FROM tennis
    WHERE (Player_1 LIKE "Nadal R." OR Player_1 LIKE "Federer R.")
    AND (Player_2 LIKE"Federer R." OR Player_2 LIKE "Nadal R.")
    ORDER BY Date ASC

)
WHERE Winner ="Federer R."
```
